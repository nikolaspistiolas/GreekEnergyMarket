$\textbf{Determenistic Forecast}$

```python
strategy2 = pd.DataFrame()
strategy2 =  Day_ahead_imbalance_prices
strategy2['Bid'] = 1.02*probabilistic_forecast['Forecast'].values
strategy2['Production'] = probabilistic_forecast['Production'].values
strategy2['Dayahead Revenue'] = strategy2['Bid']*strategy2['Day_ahead_price']

strategy2['Imabalance_Revenue'] = strategy2[columns_prices[1]]*( strategy2["Bid"] -strategy2["Production"])

strategy2['Total_Revenue'] = strategy2["Production"]*strategy2["Day_ahead_price"] - (strategy2["Day_ahead_price"]- strategy2[columns_prices[1]])*(strategy2["Production"] - strategy2["Bid"])
```

```python
strategy2_revenue_new = strategy2['Total_Revenue'].sum()
strategy2_ratio_new = strategy2_revenue_new/reference_revenue

print('The revenue for the Determenistic Forecast (Perfect Forecast): {}'.format(strategy2_revenue_new))
```

```python
strategy3_probabilistic = probabilistic_forecast
strategy3_probabilistic["Day_Ahead_price"] = Day_ahead_imbalance_prices["Day_ahead_price"].values
strategy3_probabilistic["Imbalance_price"] = Day_ahead_imbalance_prices[columns_prices[1]].values
```

```python
strategy3_probabilistic.head(50)
```

```python
##### Using Probabilistic Forecast (Risk Constrained Trading Strategy)
risk_down = strategy3_probabilistic[strategy3_probabilistic['Prediction_Random_Forest_Regression'] == 0][["q20","Day_Ahead_price","Imbalance_price","Production"]]
risk_up   = strategy3_probabilistic[strategy3_probabilistic['Prediction_Random_Forest_Regression'] == 1][["q80","Day_Ahead_price","Imbalance_price","Production"]]

##### Using Probabilistic Forecast (Risk Constrained Trading Strategy) knowing Diff Direction 
perfect_knowledge_diff_price_down = strategy3_probabilistic[strategy3_probabilistic['Diff_direction'] == 0][["q10","Day_Ahead_price","Imbalance_price","Production"]]
perfect_knowledge_diff_price_up = strategy3_probabilistic[strategy3_probabilistic['Diff_direction'] == 1][["q90","Day_Ahead_price","Imbalance_price","Production"]]

##### No Risk solution Using the maximum and 0 production for bids 
no_risk_strategy_down = strategy3_probabilistic[strategy3_probabilistic['Prediction_Random_Forest_Regression'] == 0][["Day_Ahead_price","Imbalance_price","Production"]]
no_risk_strategy_up = strategy3_probabilistic[strategy3_probabilistic['Prediction_Random_Forest_Regression']   ==1][["Day_Ahead_price","Imbalance_price","Production"]]

no_risk_strategy_down["Bid"] = 0
no_risk_strategy_up["Bid"] = 6.86

##### No Risk solution Using the maximum and 0 production for bids and by knowing the Diff Direction 

no_risk_strategy_down_pk = strategy3_probabilistic[strategy3_probabilistic['Diff_direction'] == 0][["Day_Ahead_price","Imbalance_price","Production"]]
no_risk_strategy_up_pk   = strategy3_probabilistic[strategy3_probabilistic['Diff_direction'] ==1][["Day_Ahead_price","Imbalance_price","Production"]]

no_risk_strategy_down_pk["Bid"] = 0
no_risk_strategy_up_pk["Bid"] = 6.86
```

```python
risk_down["Revenue"] = risk_down["Production"]*risk_down["Day_Ahead_price"] - (risk_down["Day_Ahead_price"]- risk_down["Imbalance_price"])*(risk_down["Production"] - risk_down["q20"])
risk_up["Revenue"] = risk_up["Production"]*risk_up["Day_Ahead_price"] - (risk_up["Day_Ahead_price"] - risk_up["Imbalance_price"])*(risk_up["Production"] -risk_up["q80"])
```

```python
perfect_knowledge_diff_price_down["Revenue"] = perfect_knowledge_diff_price_down["Production"]*perfect_knowledge_diff_price_down["Day_Ahead_price"] - (perfect_knowledge_diff_price_down["Day_Ahead_price"]- perfect_knowledge_diff_price_down["Imbalance_price"])*(perfect_knowledge_diff_price_down["Production"] - perfect_knowledge_diff_price_down["q10"])
perfect_knowledge_diff_price_up["Revenue"] = perfect_knowledge_diff_price_up["Production"]*perfect_knowledge_diff_price_up["Day_Ahead_price"] - (perfect_knowledge_diff_price_up["Day_Ahead_price"] - perfect_knowledge_diff_price_up["Imbalance_price"])*(perfect_knowledge_diff_price_up["Production"] -perfect_knowledge_diff_price_up["q90"])
```

```python
no_risk_strategy_down["Revenue"] = no_risk_strategy_down["Production"]*no_risk_strategy_down["Day_Ahead_price"] - (no_risk_strategy_down["Day_Ahead_price"]- no_risk_strategy_down["Imbalance_price"])*(no_risk_strategy_down["Production"] - no_risk_strategy_down["Bid"])
no_risk_strategy_up["Revenue"] = no_risk_strategy_up["Production"]*no_risk_strategy_up["Day_Ahead_price"] - (no_risk_strategy_up["Day_Ahead_price"] - no_risk_strategy_up["Imbalance_price"])*(no_risk_strategy_up["Production"] -no_risk_strategy_up["Bid"])
```

```python
no_risk_strategy_down_pk["Revenue"] = no_risk_strategy_down_pk["Production"]*no_risk_strategy_down_pk["Day_Ahead_price"] - (no_risk_strategy_down_pk["Day_Ahead_price"]- no_risk_strategy_down_pk["Imbalance_price"])*(no_risk_strategy_down_pk["Production"] - no_risk_strategy_down_pk["Bid"])
no_risk_strategy_up_pk["Revenue"] = no_risk_strategy_up_pk["Production"]*no_risk_strategy_up_pk["Day_Ahead_price"] - (no_risk_strategy_up_pk["Day_Ahead_price"] - no_risk_strategy_up_pk["Imbalance_price"])*(no_risk_strategy_up_pk["Production"] -no_risk_strategy_up_pk["Bid"])
```

```python
strategy3_probabilistic_revenue = risk_down['Revenue'].sum() + risk_up["Revenue"].sum()
strategy3_ratio_probabilistic = strategy3_probabilistic_revenue/reference_revenue

print('The revenue for the Probabilistic Strategy using Random Forest Regression for classification : {}'.format(strategy3_ratio_probabilistic))
```

```python
strategy3_probabilistic_perfect_knowledge = perfect_knowledge_diff_price_down['Revenue'].sum() + perfect_knowledge_diff_price_up["Revenue"].sum()
strategy3_ratio_probabilistic_perfect_knowledge = strategy3_probabilistic_perfect_knowledge/reference_revenue

print('The revenue for the Probabilistic Strategy using Random Forest Regression for classification : {}'.format(strategy3_ratio_probabilistic_perfect_knowledge))
```

```python
strategy3_probabilistic_no_risk = no_risk_strategy_down['Revenue'].sum() + no_risk_strategy_up["Revenue"].sum()
strategy3_ratio_no_risk = strategy3_probabilistic_no_risk/reference_revenue

print('The revenue for the optimal strategy with prediction using Random Forest Regression : {}'.format(strategy3_ratio_no_risk))
```

```python
##### Best Revenue Overall
strategy3_probabilistic_no_risk_pk = no_risk_strategy_down_pk['Revenue'].sum() + no_risk_strategy_up_pk["Revenue"].sum()
strategy3_ratio_no_risk_pk = strategy3_probabilistic_no_risk_pk/reference_revenue

print('The revenue for the optimal strategy : {}'.format(strategy3_ratio_no_risk_pk))
```
