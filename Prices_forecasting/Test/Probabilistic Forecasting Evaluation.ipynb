$\textbf{Λοιπές Χρεώσεις}$

* $DEV_{e,t}$ H απόκλιση για κάθε Αγοριαία Χρονική μονάδα t.
$$DEV_{e,t} = MQ_{e,t} - MS_{e,t}$$
* $ADEV_{e,m}$ η μηνιαία απόκλιση για τον μήνα m , 

$$ADEV{e,m} = \sum_{t \in m }|DEV_{e,t}|$$

* $NADEV_{e,m}$ η μηνιαία ενεργός τιμή των αποκλίσεων 

$$NADEV_{e,m} = \frac{ADEV_{e,m}}{\sum_{t\in m} MQ_{e,t}}$$

* $RMSDEV_{e,m}$ Η κανονικοποιημ'ένη ενεργός τιμή για τον μήνα m
$$RMSDEV_{e,m} = \sqrt{\sum_{t\in m } DEV_{e,t}^2}$$


* $NRMSDEV_{e,m}$ : Για την οντότητα με ευθήνη εξισορρόπησης 
$$NRMSDEV_{e,m} = \frac{RMSDEV_{e,m}}{\sqrt{\sum_{t\in m }(MQ_{e,t}^2)}}$$
